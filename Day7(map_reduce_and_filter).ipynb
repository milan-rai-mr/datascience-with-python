Reduce Function

```python
# function, class, module, package, library
```

```python
import functools
```

```python
l = [1,2,3,4,5]
```

```python
functools.reduce(lambda x,y : x+y,l)
```

Not more than two parameters and not less than one element can be passed in reduce

```python
l = [1,2,3,4]
```

```python
functools.reduce(lambda x,y,z : x+y+z,l)
```

```python
functools.reduce(lambda x,y : x+y,[]) # Error blank list
```

```python
functools.reduce(lambda a,b : a-b, [10])
```

```python
functools.reduce(lambda a,b : a/b, [20])
```

```python
functools.reduce(lambda a,b : a*b, [10,20])
```

```python
functools.reduce(lambda a,b,c,d : a*b+c+d, [20]) #Exceptional one element
```

Filter Function

```python
l
```

```python
filter(lambda x : x > 3, l)
```

```python
tuple(filter(lambda x : x > 3, l))
```

```python
list(filter(lambda x : x % 2 == 0, l))
```

```python
set(filter(lambda x : x % 2 != 0, l))
```

```python
list(filter(lambda x : x % 2 != 1, l))
```

```python
list(filter(lambda x : x % 2 == 1, l))
```

```python
l3 = [-1,2,-4,3,-3,5,-6,0]
```

```python
list(filter(lambda x : x >=0, l3))
```

```python
list(filter(lambda x : x <=0, l3))
```

```python
l4 = ['china', 'india', 'nepal', 'bhutan', 'russia', 'america']
```

```python
set(filter(lambda x : len(x) == 5,l4))
```
