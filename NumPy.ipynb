np.arange() function in NumPy works like Python's range() to create sequences of numbers. Its syntax is:
np.arange(start,end,step)
- start (optional): The starting number (default is 0).
- end: The ending number (not included in the array).
- step (optional): The difference between numbers (default is 1).

```python
array1 = range(10)
print(list(array1))

array2 = range(3, 20, 2)
print(list(array2))

array3 = range(3, 20, 2.3)
print(array3)
```

# Array Slicing

Array slicing in NumPy allows you to select parts of an array, just like slicing in Python lists.

```python
# create a numpy array
numbers = np.array([2, 4, 6, 8, 10, 12, 14])

# access the first 4 elements of the array
sliced_numbers = numbers[:4]
print(sliced_numbers)
```

```python
# create a numpy array
numbers = np.array([2, 4, 6, 8, 10, 12, 14])

sliced_numbers = numbers[1:3]
print(sliced_numbers)
```

If no start is given, it starts from the first element. If no stop is given, it goes to the end.

```python
# create a numpy array
numbers = np.array([2, 4, 6, 8, 10, 12, 14])

# access the elements at even index
sliced_numbers = numbers[0::2]
print(sliced_numbers)
```

```python
# create a numpy array
numbers = np.array([2, 4, 6, 8, 10, 12, 14])

# access the elements at even index
sliced_numbers = numbers[::3]
print(sliced_numbers)
```

Negative slicing in NumPy allows you to access elements from the end of an array using negative indices.

```python
# Negative Slicing

# create a 1D numpy array
arr = np.array([1, 2, 3, 4, 5, 6])

# slice the last three elements of the array
sliced_arr = arr[-3:]

print(sliced_arr)
```

```python
arr = np.array([1,2,3,4,5,6])

# slice the array in reverse order
sliced_arr = arr[-1::-1]
print(sliced_arr)
```

```python
# create a numpy array
numbers = np.array([2, 4, 6, 8, 10, 12])

# create reversed array
reversed_numbers = numbers[::-1]
print(reversed_numbers)
```

# Array Attribute



Array attributes are simple properties of NumPy arrays that tell us basic details about the array, like:

- Its size (number of elements).  
- Its shape (how many rows and columns it has).  
- The type of data it holds.  
- The number of dimensions it has (1D, 2D, 3D, etc.).  

These attributes help us understand and work with the array easily.

**Attributes and their meanings:**
- **ndim**: Number of dimensions (eg. 1 for 1D array).
- **shape**: Size of the array in each dimension (e.g., length for 1D).
- **size**: Total number of elements in the array.
- **dtype**: Type of data stored in the array.
- **itemsize**: Size of each elements in the bytes.
- **nbytes**: Total memory used by the array (itemsize x size)

```python
import numpy as np
```

```python
# create a numpy array
numbers = np.array([2,4,6,8,10,12,14])

# give the array dimension
print('Number of dimensions:', numbers.ndim)

# get shape of the array
print('Shape:', numbers.shape)
```

```python
numbers1 = np.array([2,4,6,8,10,'a'])
print(numbers1)
print('Data Types of First Array:', numbers1.dtype)

# create an array of int type
int_num = np.array([2,4,6,8,10,12,13])

# creat an array of float type
float_num = np.array([3.4,5.6,7.8,9.1])

# check data type of numbers
print('Data Types of Second Array:', int_num.dtype)

# check data type of float array
print('Data Types of Third Array:', float_num.dtype)
```

We use size attribute to find the total number of elements in the array.

```python
first = np.array([1,3,5,7,8,9,12])
second = np.array([1,2,4,6,10])

print('Size of first:', first.size)
print('Size of second:', second.size)
```

We use itemsize attribute to find the size (in bytes) of each element in the array.

```python
# create a default in 1-D array of integers
array1 = np.array([6,7,8,9,10,11])

# create a 1-D array of 32-bit integers
array2 = np.array([6,7,8,9,10,11], dtype=np.int32)

# create a 1-D array of floats
array3 = np.array([1,2,2.4,5.6,7.8], dtype=np.float128)

print(array1)
print(array2)
print(array3)

print(array1.itemsize)
print(array2.itemsize)
print(array3.itemsize)
```

nbytes finds the total number of bytes consumed by elements in the array

```python
print(array1.nbytes)
print(array2.nbytes)
print(array3.nbytes)
```

```python
array4 = np.array([1,2,3,4,5], dtype = np.float32)
print(array4)
print(array4.itemsize)
print(array4.nbytes)
```
