Casefold vs. Lower
•	Create two strings, "Case" and "case".
•	Compare them with the regular == operator directly.
•	Compare them again after applying .casefold().
•	Print results and comment on how .casefold() differs from .lower() in edge cases.


```python
a= "Case"
b="case"
```

```python
a == b
```

```python
str.casefold(a) == b
```

```python
a.lower() == b
```

For basic English, casefold() and lower() give the same result. The difference apperas with special characters - e.g. German language  'ß'.casefold() gives 'ss' but 'ß'.lower() stays 'ß'. For International text, casefold() is safer.
