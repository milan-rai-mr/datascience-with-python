function = lambda inputs : returning expression

```python
pow = lambda n,p : n ** p
```

```python
pow(2,3)
```

```python
pow(4,2)
```

```python
power = lambda n,p : n**p
```

```python
power(2,3)
```

```python
power(5,8)
```

```python
def max_value(x,y):
    if x > y:
        return x
    else:
        return y

result = max_value(44, 67)
result
```

```python
max_value(3,55)
```

```python
max_value(112,11.2)
```
