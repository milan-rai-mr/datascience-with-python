```python
def number(a):
  l1 = []
  for i in a:
    if type(i) == list or type(i) == tuple or type(i) == set:
      for j in i:
        if type(j) == int or type(j) == float:
          l1.append(j)
    else:
      if type(i) == int or type(i) == float:
        l1.append(i)
  return l1
```

```python
number([1,2,3,[4,5]])
```

```python
def numeric_data(a): #docstring
  '''function to extract numeric data from list'''
  l1 = []
  for i in a:
    if type(i) == list or type(i) == tuple or type(i) == set:
      for j in i:
        if type(j) == int or type(j) == float:
          l1.append(j)
    else:
      if type(i) == int or type(i) == float:
        l1.append(i)
  return l1
```

```python
numeric_data()
```

```python
numeric_data(l)
```

```python
numeric_data([4,5,6,'a','b'])
```
