While loop

While Loop is used when we need to run specific block of code until a condition is met.

```python
'''
  while condition:
    // code block to be executed
'''
```

when the condition has < , use +

and when the condition has > , use -

```python
a = 1
while a <= 4:
  print(a)
  a = a + 1
```

```python
a = 5
while a >= 1:
  print(a)
  a = a - 1
```

```python
#sum of the number till some point
```

```python
n = int(input("Enter a number: "))
sum = 0
counter = 1

while counter <= n:
  sum = sum + counter
  counter = counter + 1
print('Total sum is: ' , sum)
```

```python
#factorial
```

```python
number = int(input('Enter a positive number: '))
factorial = 1
while(number >= 1):
  factorial = factorial * number
  number = number -1
print(f'The factorial is {factorial}')
```

```python
num = int(input('Enter the element: '))
a,b = 0,1
counter = 0
while counter < num:
  print(a)
  c = a + b
  a = b
  b = c
  counter = counter + 1
```

```python
word = input("Enter the string for reversal: ")
reverse = ''
length = len(word)
while length > 0:
  reverse = reverse + word[length - 1]
  length = length - 1
print(reverse)
```

```python
n = int(input('Enter number: '))
i = 1
while(i <= 10):
  result = i * n
  print(f'{n} * {i} = {result}')
  i = i + 1
```

```python
n = int(input('Enter number: '))
i = 1
q = int(input(f'Enter the number upto which you want multiplication: '))
while(i <= q):
  result = n * i
  print(f'{n} * {i} = {result}')
  i = i + 1
```

```python
n = int(input('Enter number: '))
i = 1
q = int(input(f'Enter the number upto which you want multiplication: '))
while(i <= q):
  result = n * i
  print(n, "*", i, "=", result)
  i = i + 1
```
