While loop

While Loop is used when

```python
n = int(input("Enter a number: "))
sum = 0
counter = 1

while counter <= n:
  sum = sum + counter
  counter = counter + 1
print('Total sum is: ' , sum)
```

```python
word = input("Enter the string for reversal: ")
reverse = ''
length = len(word)
while length > 0:
  reverse = reverse + word[length - 1]
  length = length - 1
print(reverse)
```
