```python
data = {
    'name' : 'rajesh',
    'mail_id' : 'rajesh@gmail.com',
    'phone_no' : 861616161,
    'subject' : ['data science', 'web development', 'data analysis']
}
```

```python
type(data)
```

```python
import json
```

```python
with open('data.json','w') as f:
  json.dump(data,f)
  f.close()
```

```python
with open('data.json','r') as f:
  data1 = json.load(f)
```

```python
import csv
```

```python
data = [
    ['name','email_id','phone_no'],
    ['nabin','nk@gmail.com',987654],
    ['milan','m@gmail.com',981234],
    ['sagar','s@gmail.com',980000]
]
```

```python
with open('data.csv','w') as f:
  writer = csv.writer(f)

  for i in data:
    writer.writerow(i)
```

```python
with open('data.csv','r') as f:
  read_data = csv.reader(f)

  for i in read_data:
    print(i)
```
