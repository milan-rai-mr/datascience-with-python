```python
integer = []
decimal = []
boolean = []
strings = []
complexes = []
l = [1,2,3,4.55,'a',4-8j,True, False, True, 'b', 'c', 3+4j]
for i in l:
  if type(i) == int:
    integer.append(i)
  elif type(i) == float:
    decimal.append(i)
  elif type(i) == bool:
    boolean.append(i)
  elif type(i) == str:
    strings.append(i)
  elif type(i) == complex:
    complexes.append(i)
```
