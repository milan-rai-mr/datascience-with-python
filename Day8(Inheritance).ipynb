- Inheritance lets a new class (child class) use the properties and methods of an existing class (parent class).
- A child class is created by passing the parent class as an argument when defining it.
- Parent Class (Base Class) and Child Class (Derived Class)

Single Inheritance

```python
# Single Inheritance
# Parent Class/ Base Class
class parent:
  def parent_method(self):
    return 'this is the example of single inheritance'
```

```python
p = parent()
```

```python
p.parent_method()
```

```python
# Child Class/ Derived Class
class child(parent):
  pass
```

```python
c = child()
```

```python
c.parent_method()
```

```python
class A:
  def __init__(self,a): #constructor
    self.a = a

  def valueofa(self):
    return self.a
```

```python
class B(A):
  def __init__(self,b):
    self.b = b

  def valueofb(self):
    return self.b
```

```python
a = A(20)
```

```python
a.valueofa()
```

```python
b = B(30)
```

```python
b.valueofb()
```

Multi-Level Inheritance

```python
# Multi-Level
class class1:
  def test_class1(self):
    return 'this is a method from class1'
```

```python
class class2(class1):
  def test_class2(self):
    return 'this is a method from class2'
```

```python
cls = class2()
```

```python
cls.test_class1()
```

```python
cls.test_class2()
```

```python
class class3(class2):
  pass
```

```python
obj_class1 = class1()
```

```python
obj_class1.test_class1()
```

```python
obj_class2 = class2()
```

```python
obj_class2.test_class1()
```

```python
obj_class2.test_class2()
```

```python
obj_class3 = class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```

Multiple Inheritance

```python
# Multiple Inheritance
class father:
  def father_property(self):
    return 'this is the property of father'
```

```python
class mother:
  def mother_property(self):
    return 'this is the property of mother'
```

```python
class child(father,mother):
  pass
```

```python
obj_child = child()
```

```python
obj_child.father_property()
```

```python
obj_child.mother_property()
```

Hierarchical

```python
#Hierarchical
class Animal:
  def sounds(self):
    return "Animals make sounds"

class Dog(Animal):
  def sound(self):
    return "Dog barks"

class Cat(Animal):
  def sound(self):
    return "Cat meows"

class Goat(Animal):
  def sound(self):
    return "Goat bleats"

dog = Dog()
cat = Cat()
goat = Goat()
```

```python
dog.sound()
```

```python
dog.sounds()
```

```python
cat.sound()
```

```python
cat.sounds()
```

```python
print(goat.sounds())
print(goat.sound())
```
