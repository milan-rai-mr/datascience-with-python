```python
d = {
    'key': 'nepal',
     1 : [2,3,4]
}

print(d['key1'])
```

```python
try:
  d = {
    'key': 'nepal',
     1 : [2,3,4]
  }

  print(d['key2'])
except KeyError as e:
  print(e)
```

```python
try:
  with open('text.txt','r') as f:
    f.read()
except FileNotFoundError as e:
  print(e)
```

```python
try:
  with open('test.txt','r') as f:
    f.read()
except Exception as e: # generic must not be in top, it must be at bottom
  print(e)
except FileNotFoundError as e:
  print('test',e)
```

```python
try:
  with open('test.txt','r') as f:
    f.read()
except FileNotFoundError as e:
  print('test',e)
except Exception as e: # generic must not be in top, it must be at bottom
  print(e)
```
