- A **for loop** in Python is used to iterate over a sequence (like a list, string, or range).
- It starts with the keyword `for`, followed by a variable name and the sequence (e.g., `for i in range(5):`).
- The loop runs for each item in the sequence, with the variable taking the value of each item in turn.
- Inside the loop, we can perform any actions or operations we want for each value.
- This helps us to avoid repeating code when we need to do something multiple times for each item in a sequence.

```python
#
'''
  for variable in iterable:
    Code block to be executed
'''
```
