Abstraction in Python's Object-Oriented Programming(OOP) means hiding unnecessary details and showing only the relecant parts of an object to the user. It focuses on the what rather than the how.


- Abstraction focuses on hiding implementation details and showing only the essential features of an object.
- Encapsulation focuses on hiding data by restricting access to certain compenents.

```python
class tu:
  def student_details(self):
    pass

  def student_assignment(self):
    pass

  def student_marks(self):
    pass
```

```python
import abc
class tu:

  @abc.abstractmethod
  def student_details(self):
    pass

  @abc.abstractmethod
  def student_assignment(self):
    pass

  @abc.abstractmethod
  def student_marks(self):
    pass
```

```python
class student_details(tu):
  def student_details(self):
      return 'this is a method for taking student details'

  def student_assignment(self):
      return 'this is a method for assignment details for a particular student'
```

```python
class data_analysis(tu):
  def student_details(self):
    return 'this will return the details for data analysis course'
  def student_assignment(self):
    return 'this will give you assignment details for data analysis course'
```

```python
da = data_analysis()
da.student_details()
```

```python
sd = student_details()
sd.student_details()
```

```python
da.student_assignment()
```

```python
sd.student_assignment()
```
