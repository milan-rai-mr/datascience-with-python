Polymorphism means "many forms". It allows the same function or method to work differently depending on the object that calls it.

In polymorphism, methods(functions) with the same name can perform different tasks based on the object. Example: A fly() method for a Bird class might mean "flap wings", while for an Airplane class, it means "start engines".

Think of the word "run". A person can run (move quickly). A computer program can run (execute commands). A car engine can run (operate). The action depends on who/what is performing it, just like polymorphism.

```python
class ai:
    def syllabus(self):
      print('python, numpy, pandas, matplotlib, seaborn, sql, tableau, scikit-learn,pytorch,langchain ')
```

```python
class mern:
    def syllabus(self):
      print('html, css, javasctipt, mongodb, express, react, node')
```

```python
d = ai()
```

```python
m = mern()
```

```python
d.syllabus()
```

```python
m.syllabus()
```

```python
def classes(class_obj):
  for i in class_obj:
    i.syllabus()
```

```python
ai = ai()
```

```python
mern = mern()
```

```python
obj = [ai,mern]
```

```python
classes(obj)
```
