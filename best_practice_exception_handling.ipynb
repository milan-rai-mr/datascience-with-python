```python
# always use a specific exception
try:
  10/0
except Exception as e:
  print(e)
```

```python
try:
  10/0
except ZeroDivisionError as e:
  print(e)
```

```python
# always print a proper message
try:
  10/0
except ZeroDivisionError as e:
  print('I am trying to handle a zerodivision error:', e)
```

```python
# always avoid writing multiple exception handling
try:
  100/0
except FileNotFoundError as e:
  print('this is my file not found error',e)
except AttributeError as e:
  print('this is my attribute error',e)
except ZeroDivisionError as e:
  print('this is my zero division error',e)
```
