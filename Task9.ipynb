•	Write a Python function using def to return the product of any count of numbers that is passed as an argument to the function.

•	Write a function that takes different objects (Rectangle, Circle, Square) and calls their area() method.

```python
import math
class Rectangle:
  def __init__(self,length,breadth):
    self.length = length
    self.breadth = breadth

  def area(self):
    return self.length * self.breadth

class Circle:
  def __init__(self,radius):
    self.radius = radius
  def area(self):
    return math.pi * self.radius ** 2

class Square:
  def __init__(self,side):
    self.side = side
  def area(self):
    return self.side * self.side

def show_area(shape):
    print(shape.area())

show_area(Rectangle(2,10))
show_area(Circle(10))
show_area(Square(2))
```

•	Use the ABC module to define an abstract class Shape with an abstract method calculate_area() and implement it in Circle and Rectangle subclasses.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius
  def calculate_area(self):
    return math.pi*self.radius**2

class Rectangle(Shape):
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth
  def calculate_area(self):
    return self.length * self.breadth

c = Circle(7)
r = Rectangle(5,10)
print(c.calculate_area())
print(r.calculate_area())
```

•	Implement a function that raises a CustomException when the user enters a odd number the exception is raised.

```python
class CustomException(Exception):
  pass

def check_odd(num):
  if num %2 != 0:
    raise CustomException("The number is odd.")

try:
  number = int(input("Enter a number: "))
  check_odd(number)
except CustomException as e:
  print(e)
```
