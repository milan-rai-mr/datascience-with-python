```python
class ValidateAge(Exception):
  def __init__(self,msg):
    self.msg = msg
```

```python
def validateage(age):
  if age < 0:
    raise ValidateAge('Entered age is negative')
  elif age > 125:
    raise ValidateAge('Entered age is very high')
  else:
    print('Age is valid')
```

```python
validateage(44)
```

```python
validateage(-22)
```

```python
validateage(139)
```

```python
try:
  age = int(input('Enter your age:'))
  validateage(age)
except ValidateAge as e:
  print(e)
```
