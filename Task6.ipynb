Print Numbers from 10 to 30

```python
a = 1
while a <= 20:
  a = a + 1
  if a %2 != 0:
    print(a)

```

Keep Asking for a Number Until It's Positive
•	Keep asking the user to enter a number until they enter a positive number.


```python
user = int(input("Enter a positive number: "))

while user < 0:
  user =  int(input("Please enter a positive number only."))

print("the positive number: ", user)
```

Guess the Secret Number using while loop
•	Secret number = 7.
•	Ask the user to guess the number until it's correct.


```python
user = int(input("Enter the secret number: "))

while user != 7:
  user = int(input("Try Again: "))
else:
  print("You got it right.")
```

Function to Check Even or Odd
•	Create a function check_even(num) that prints "Even" or "Odd".


```python
def check_even(num):
  if num %2 == 0:
    print('Even')
  else:
    print('Odd')

```
