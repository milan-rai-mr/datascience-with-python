Tuple


*   List is mutable and Tuple is immutable.
*   List is a group of elements enclosed inside [] whereas Tuple is a group of elements enclosed inside ().







*   Tuple in Python is like a list, but we cannot change its items once it is created
*   Tuple is written using round brackets(), with items seperated by commas, like this: my_tuple = (1, 2, "banana", 4)
*  they are useful when we want to store a group of items that should stay the same and not be modified.






```python
tpl = (1, 2, 3, 4, 5, 6, 7, 8, 9)
```

```python
type(tpl)
```

```python
tpl[4]
```

```python
tpl[2:7:2]
```

Set


*   Set in Python is a collection of items where each item is unique, meaning no duplicates are allowed.
*   Set is written using curly brackets{} containing at least one element, like this: my_set = {1, 2, 3, "apple"}


*   The items in a set are unordered, so they don't have a specific position, and we cant access them using an index like in a list.








*   We can do indexing and slicing in list and tuple,  but we can't do indexing and slicing in set
*   Duplicate data are allowed in list and tuple but not in set



The elements in the set should not be mutable.
