OOPS (Object Oriented Programming System)

- Object-Oriented Programming is a way of designing and organizing code by using "objects." These objects represent real-world things like a car, a dog, or a book, and they have attributes (data) and behaviors (functions).
- A class is like a blueprint for creating objects. For example, a "Car" class defines what attributes (e.g., color, model) and behaviors (e.g., drive, stop) a car can have.
An object is a specific instance of a class, like a red Tesla Model 3.
- Attributes are the data stored in an object (e.g., a car's color is "red").
Methods are the functions that an object can perform (e.g., a car can "drive" or "stop").
- OOP makes code reusable, organized, and easier to understand. It helps manage complexity in large programs by breaking them into smaller, manageable parts.

```python
#__init__ constructor
class Car:
  def __init__(self,brand,color):
    self.brand = brand
    self.color = color
```

```python
car1 = Car('Toyota','Red')
car2 = Car('BMW','Black')
```

```python
print(car1.brand) # self.brand = car1.brand
print(car2.color) # self.color = car2.color
```

```python
# Methods are functions inside a class
class Car:
  def __init__(self,brand): # construcor
    self.brand = brand

  def drive(self):  # method
    print(self.brand, "is driving")
```

```python
car = Car("Tesla")
car.drive()
```

```python
car.brand
```

```python
class tu: # init used to pass data specific to an object inside the class and work accoring to it
  def __init__ (self, phone_no, email_id, student_id): # constructor
    self.phone_no = phone_no
    self.email_id = email_id
    self.student_id = student_id
  def return_student_details(self):
    return self.phone_no, self.email_id, self.student_id
```

```python
sagar = tu(9876543210, 'sagar@gmail.com', 1)
```

```python
sagar.phone_no
```

```python
sagar.email_id
```

```python
sagar.student_id
```

```python
sagar.return_student_details()
```
