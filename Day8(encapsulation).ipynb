- Encapsulation means hiding the details of an object and only exposing what is necessary.
- It keeps the data(attribute) and functions(methods) inside a class safe from being directly accessed or modified.
- Public: Data or methods that anyone can access(normal_variable).
- Private: Data or methods that only the clas can access(__private_variable).
- Protected: Data or methods that can be accessed by the class and its subclasses(_protected_variable).
- For eg: We can deposit or withdraw money, but you can't directly access or change the balance. The bank provides secure methods to interact with the amount.

```python
class test:
  def __init__(self,a,b):
    self.a = a
    self.b = b
```

```python
t = test(1000,3000)
```

Encapsulation helps to keep entities private so other cannot see internal structure and modify

```python
class car: # __ makes variable private which can't be directly accessed
  def __init__(self,year,make,model,speed=0):
    self.__year = year
    self.__make = make
    self.__model = model
    self.__speed = speed
```

```python
c = car(2024, 'toyota','innova',12)
```

```python
c.year
```

```python
c.__year
```

```python
c._car__year
```

```python
c._car__year = 1223
```

```python
c._car__year
```

```python
class car: # __ makes variable private which can't be directly accessed
  def __init__(self,year,make,model,speed):
    self.__year = year
    self.__make = make
    self.__model = model
    self.__speed = speed

  def set_speed(self,speed):
    self.__speed = 0 if speed <= 0 else speed

  def get_speed(self):
    return self.__speed
```

```python
c = car(2024, 'toyota','innova',12)
```

```python
c._car__speed
```

```python
c.set_speed(100)
```

```python
c.get_speed()
```

```python
c._car__speed
```

```python
class bank_account:
  def __init__(self,balance):
    self.__balance = balance

  def deposit(self, amount):
    if amount <= 0:
        return 'Invalid Deposit'
    self.__balance = self.__balance + amount
    # self.balance += amount
    return f'Rs.{amount} Depostied and final balance after deposit is Rs.{self.__balance}'

  def withdraw(self, amount):
    if self.__balance >= amount:
        self.__balance = self.__balance - amount
        # self.__balance -= amount
        return f'Rs.{self.__balance} left in bank account after Rs.{amount} is withdrawn. The total amount in the bank before withdraw was Rs.{self.__balance + amount}'
    else:
        return 'Invalid Withdrawl. Since the required amount is not in the account.'

  def get_balance(self):
    return f'The total balance in your account is Rs.{self.__balance}'
```

```python
diwas = bank_account(1000)
```

```python
diwas.get_balance()
```

```python
diwas.withdraw(200)
```

```python
diwas.deposit(1000)
```
