Exception Handling
- It is a way to handle errors in the program without crashing it.
- Instead of the program stopping when an error occurs, exception handling lets us control what happens next.
- Use try to write the code that might cause an error.
- Use except to define what to do if an error happens.
- else: Runs if no error occurs.
- finally: Runs no matter what (even if there’s an error).

```python
a = 10
```

```python
a / 10
```

```python
a / 0
```

```python
a = 6
try:
  print(a/0)
except Exception as e:
  print(e)
```
